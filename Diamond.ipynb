### Data Visualization on Diamond Data set

```python
import seaborn as sns
```

```python
diamond = sns.load_dataset("diamonds")
diamond
```

```python
sns.histplot(data=diamond, x="carat", kde=True)
```

Insights:-
  The histogram with KDE suggests that carat weights between 0.2 and 1.0 are most common, with a smaller peak around 1.5 carats.

```python
sns.jointplot(data=diamond, x="carat", y="price", kind="scatter")
```

insights:
plot shows a positive correlation between carat and price, indicating that as carat weight increases, diamond price tends to increase.

```python
sns.jointplot(data=diamond, x="carat", y="price", kind="hex")
```

insights:
plot reveals the relationship between carat and price, with denser areas indicating where carat and price values cluster. It highlights the positive correlation between carat weight and diamond price.

```python
sns.jointplot(data=diamond, x="carat", y="price", kind="reg")
```

insights:
plot displays a linear regression line, showing a clear positive relationship between carat weight and diamond price. As carat weight increases, the expected price also increases, following a linear trend.

```python
sns.jointplot(data=diamond, x="carat", y="price", hue='cut')
```

Insights:
    plot provides a visual representation of the relationship between carat weight and price, with points colored by different diamond cut qualities. It allows for a comparative analysis of how different cut qualities impact the carat vs. price relationship, showing variations in distribution and correlation patterns among cut categories.

```python
sns.kdeplot(data=diamond)
```

```python
sns.kdeplot(data=diamond, x="price")
```

Insights:
plot shows the distribution of diamond prices, revealing multiple peaks or clusters, suggesting different price categories or groups within the dataset.

```python
pair = ["carat", "depth", "table", "price"]
sns.pairplot(data=diamond[pair])
```

Insights:
Carat vs. Depth: No strong linear relationship; carat values evenly distributed across different depths.
Carat vs. Table: No strong linear relationship; carat values vary across various table widths.
Carat vs. Price: Positive correlation; as carat weight increases, diamond price tends to rise.
Depth vs. Table: No clear linear relationship; data points scattered across different depth and table combinations.
Depth vs. Price: No strong linear relationship; diamond prices vary with different depth measurements.
Table vs. Price: No strong linear relationship; diamond prices vary across different table measurements.
